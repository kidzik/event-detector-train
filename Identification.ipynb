```python
import os
import numpy as np
import math
import keras.backend as K
import matplotlib.pyplot as plt

from scipy.ndimage.filters import gaussian_filter
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM, TimeDistributed, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from itertools import product
from random import randint
from scipy import signal
from peakdet import peakdet
```

For each trial per leg $r$ we have $n_r$ sampling points of $15$ joint angles ($5$ joints, $3$ DOF). Input observations are stored as matrices $X_r \in \mathbb{R}^{n_r \times 15}$. For each trial we have some of the events (foot strike / toes off) annotated. Let's focus on the foot strike event. We encode these events as a binary vector $Y_r \in \{ 0, 1\}^{n_r}$. The task is to predict these events.

For training the network we will use fixed length sequences. We set the length to `nseqlen` and use only samples from the first event `nstart - nrand` where `nstart` is the first nonzero element in $Y_r$ and `nrand` is a random integer between `nseqlen / 2`. We do this subtraction in order to simulate starting from different points - to make the procedure robust to various setups.

```python
# Settings
fdir = "csv"
files = os.listdir(fdir)
input_dim = 15
output_dim = 2

nseqlen = 128

# Merge inputs from different files together
inputs = np.zeros((len(files), nseqlen, input_dim))
outputs = np.zeros((len(files), nseqlen, output_dim))

n = 0
```

```python
for i,filename in enumerate(files):
    # print("Processing %s..." % (filename,))
    R = np.loadtxt("%s/%s" % (fdir, filename), delimiter=',')

    # find first event
    positives1 = np.where(R[:,input_dim] > 0.5)
    positives2 = np.where(R[:,input_dim + 1] > 0.5)
    if len(positives1) + len(positives2) == 0:
        continue
    nstart = max(positives1[0][0], positives2[0][0])
    nstart = nstart - randint(0,nseqlen / 2)

    if R.shape[0] < (nstart + nseqlen):
        continue

    X = R[nstart:(nstart + nseqlen),0:input_dim]
    Y = R[nstart:(nstart + nseqlen),input_dim:(input_dim + output_dim)]
    
    # Y = gaussian_filter(Y * 1.0, 1.0)

    if (not Y.any()):
        continue
    
    inputs[n,:,:] = X
    outputs[n,:,:] = Y.astype(int)[:,0:output_dim]
    n = n+1
    
    if n >= 10000:
        break
```

Now `inputs` contains $10000$ matrices $X \in \mathbb{R}^{\verb|nseqlen| \times 15}$ and `outputs` are corresponding output vectors $Y_r \in \{ 0, 1\}^{\verb|nseqlen|}$.

Since we have an inbalanced output vector (events are sparse, i.e. not many `1`'s in $Y_r$), we use a weighted binary crossentropy, penalazing $100$ times more for misclassified `1`'s.

```python
def weighted_binary_crossentropy(y_true, y_pred):
    
#    a1 = K.mean(K.binary_crossentropy(y_pred[0:1,:], y_true[0:1,:]), axis=-1)
#    a2 = K.mean(K.binary_crossentropy(y_pred[1:2,:], y_true[1:2,:]), axis=-1)
    a1 = K.mean(np.multiply(K.binary_crossentropy(y_pred[0:1,:], y_true[0:1,:]),(y_true[0:1,:] + 0.01)), axis=-1)
    a2 = K.mean(np.multiply(K.binary_crossentropy(y_pred[1:2,:], y_true[1:2,:]),(y_true[1:2,:] + 0.01)), axis=-1)
    return a1 + a2
```

We construct a neural network of the following architecture:
* $15$ dimensional time series as an input
* LSTM layer mapping the input to a $16$ dimensional time series
* LSTM layer mapping the previous layer to $1$ dimensional time series

```python
# Build the model
nhidden = 32
model = Sequential()
model.add(LSTM(input_shape = (input_dim,),input_dim=input_dim, output_dim=nhidden, return_sequences=True))
model.add(LSTM(output_dim = nhidden, return_sequences=True))
#model.add(LSTM(output_dim = nhidden, return_sequences=True))
model.add(TimeDistributed(Dense(output_dim, activation='sigmoid')))
model.compile(loss=weighted_binary_crossentropy, optimizer='adam', metrics=['accuracy'])
```

We train the model on $90\%$ of the data. $100$ epochs already give reasonable accuracy. 

```python
# Fit on the training set
ntrain = int(math.floor(n * 0.9))
model.fit(inputs[0:ntrain,:,:], outputs[0:ntrain,:,:], nb_epoch=100, batch_size=32, verbose=1)
```

The model outputs values which can be interpreted as likelihood of an event.

```python
# Test on the test set
scores = model.evaluate(inputs[ntrain:n,:,:], outputs[ntrain:n,:,:])
print("Test set loss: %.4f" % (scores[0]))

res = model.predict(inputs[ntrain:n,:,:])
```

Given the likelihoods we can approximate positions of peaks by finding local maxima. Below are some examples on the test set (still with fixed-length sequences). There is one plot per trail. Blue curve is the output of the neural network, grean peaks are the events, blue vertical lines are estimates of the events.

```python
%matplotlib inline
for ntrial in range(30):
    r = res[ntrial,:,:]

    peakind = peakdet(r[:,0],0.5)
    print(peakind[0])
    for k,v in peakind[0]:
        plt.axvline(x=k)
    peakind = peakdet(r[:,1],0.5)
    print(peakind[0])
    for k,v in peakind[0]:
        plt.axvline(x=k)

    plt.plot(res[ntrial,:,:])
    plt.plot(outputs[ntrain + ntrial,:,:])
    plt.show()
```

Now, we try prediction on variable length sequences. Note, that we have only few annotations per trial, while there should be more, in periodic manner. It seems it's caputred by LSTM.

```python
print(np.where( np.array([9,2,2,5,2]) > 3 ))

def peak_cmp(annotated, predicted):
    dist = 0
    if len(predicted) == 0:
        return -1
    
    for a in annotated:
        dist += min(np.abs(predicted - a))
    return dist / float(len(annotated))
```

```python
%matplotlib inline

sdist = []
for i,filename in enumerate(files):
    # print("Processing %s..." % (filename,))
    R = np.loadtxt("%s/%s" % (fdir, filename), delimiter=',')

    X = R[:,0:input_dim]
    Y = R[:,input_dim:(input_dim + output_dim)]
    r = model.predict(X.reshape((1,-1,15)))[0]
    
    peakind = peakdet(r[:,0],0.5)
    for k,v in peakind[0]:
        plt.axvline(x=k)
    
    sdist.append(peak_cmp(np.where(Y[:,0] > 0.5)[0], [k for k,v in peakind[0]]))
        
    peakind = peakdet(r[:,1],0.5)
    for k,v in peakind[0]:
        plt.axvline(x=k)

    sdist.append(peak_cmp(np.where(Y[:,1] > 0.5)[0], [k for k,v in peakind[0]]))

    plt.plot(r)
    plt.plot(Y)
#    plt.show()
    
    if i >= 100:
        break
```
